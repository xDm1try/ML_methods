```python
class MyQuadraticDiscriminantAnalysis:

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.mean_vectors = []
        self.cov_matrices = []
        for cl in self.classes:
            X_class = X[y == cl]
            mean_vector = np.mean(X_class, axis=0)
            cov_matrix = np.cov(X_class, rowvar=False)
            self.mean_vectors.append(mean_vector)
            self.cov_matrices.append(cov_matrix)
    
    def predict(self, X):
        predictions = []
        for x in X:
            class_probs = []
            for mean_vector, cov_matrix in zip(self.mean_vectors, self.cov_matrices):
                diff = x - mean_vector
                class_prob = 1 / np.sqrt(np.linalg.det(cov_matrix)) * np.exp(-0.5 * np.dot(np.dot(diff.T, np.linalg.inv(cov_matrix)), diff))
                class_probs.append(class_prob)
            predicted_class = np.argmax(class_probs)
            predictions.append(predicted_class)
        return np.array(predictions)
```
